```python
def calculate_mean_embeddings(sequences, max_length, tokenizer, model, device = "cpu"):
    tokens_ids = tokenizer.batch_encode_plus(sequences, return_tensors="pt",\
                                             padding="max_length", max_length=max_length,\
                                             truncation = True)["input_ids"].to(device)

    # Compute the embeddings
    attention_mask = tokens_ids != tokenizer.pad_token_id
    model.to(device)
    tokens_ids.to(device)
    attention_mask.to(device)
    torch_outs = model(
        tokens_ids,
        attention_mask=attention_mask,
        encoder_attention_mask=attention_mask,
        output_hidden_states=True
    )
    # Compute sequences embeddings
    embeddings = torch_outs['hidden_states'][-1].cpu().detach().numpy()
    attention_mask = torch.unsqueeze(attention_mask, dim=-1)
    mean_sequence_embeddings = torch.sum(attention_mask*embeddings, axis=-2)/torch.sum(attention_mask, axis=1)
    return mean_sequence_embeddings
```

```python
def pca(X, n_component = 2, scale = True):
    if scale:
        scaler = StandardScaler()
        X = scaler.fit_transform(X)
    pca_reducer = PCA(n_component)
    X = pca_reducer.fit_transform(X)
    return X

def umap_reduce(X, n_component = 2, scale = False):
    if scale:
        scaler = StandardScaler()
        X = scaler.fit_transform(X)
    umap_reducer = umap.UMAP()
    X = umap_reducer.fit_transform(X)
    return X


def tokenize_function(tokenizer):
    def f(examples):
        outputs = tokenizer(examples["data"])
    return f

# Define the metric for the evaluation using the f1 score
def compute_metrics_f1_score(eval_pred):
    """Computes F1 score for binary classification"""
    predictions = np.argmax(eval_pred.predictions, axis=-1)
    references = eval_pred.label_ids
    r={'f1_score': f1_score(references, predictions)}
    return r
```

```python
def group_shifted_seqs(seq_dict):
    seq_clust_alls = []
    for seq, index in seq_dict.items():
        seq_clust = []
        pad_length = int(0.1 * len(seq))
        truncated_seq = seq[pad_length: len(seq) - pad_length]
        for seq_compare, index in seq_dict.items():
            found = seq_compare.find(truncated_seq)
            if found > -1:
                seq_clust.append(index)
        seq_clust_alls.append(seq_clust)
    return seq_clust_alls

def choose_longest_seq(seq_clusts, seq_dict):
    reversed_dict = {value: key for key, value in seq_dict.items()}
    minimal_seqs = []
    indices = []
    for clust in seq_clusts:
        max_i = np.argmax([len(reversed_dict[i]) for i in clust])
        minimal_seqs.append(reversed_dict[clust[max_i]])
        indices.append(clust[max_i])
    return minimal_seqs, indices

def remove_duplicates(list_of_lists):
    set_of_tuples = set()
    unique_lists = []
    for sublist in list_of_lists:
        tuple_sublist = tuple(sublist)
        if tuple_sublist not in set_of_tuples:
            set_of_tuples.add(tuple_sublist)
            unique_lists.append(sublist)
    return unique_lists

def compress_seqs(seqs):
    curr_nums = [len(seqs)]
    minimal_seqs = seqs
    minimal_seqs = {i:s for s, i in enumerate(minimal_seqs)}
    while (len(curr_nums) < 2) or curr_nums[-1] != curr_nums[-2]:
        grouped_seq = group_shifted_seqs(minimal_seqs)
        seq_clusts = remove_duplicates(grouped_seq)
        seqs_only, indices = choose_longest_seq(seq_clusts, minimal_seqs)
        minimal_seqs = {i:s for s, i in tuple(zip(indices, seqs_only))}
        curr_nums.append(len(minimal_seqs))
        print("Reduce from {} to {}".format(curr_nums[-2], curr_nums[-1]))
    return seqs_only, indices
```

```python
minimal_he_compactors = compress_seqs(all_he_compactors)
minimal_le_compactors = compress_seqs(all_le_compactors)
```

```python
config = BertConfig.from_pretrained("zhihan1996/DNABERT-2-117M")
pretrained_dnabert_model = AutoModel.from_config(config)
dna_bert_classification_model = AutoModelForSequenceClassification.from_config(config)
tokenizer_dnabert = AutoTokenizer.from_pretrained("zhihan1996/DNABERT-2-117M", trust_remote_code=True)
```

```python
def tokenize_function(examples):
    outputs = tokenizer_dnabert(examples["data"])
    return outputs
```

```python
# Creating tokenized entropy dataset
tokenized_datasets_train_entropy = ds_train_entropy.map(
    tokenize_function,
    batched=True,
    remove_columns=["data"],
)
tokenized_datasets_validation_entropy = ds_validation_entropy.map(
    tokenize_function,
    batched=True,
    remove_columns=["data"],
)
tokenized_datasets_test_entropy = ds_test_entropy.map(
    tokenize_function,
    batched=True,
    remove_columns=["data"],
)
```

```python
batch_size = 16
model_name='DNA-BERT-2'
args_entropy = TrainingArguments(
    f"{model_name}-finetuned",
    remove_unused_columns=False,
    evaluation_strategy="steps",
    save_strategy="steps",
    learning_rate=1e-5,
    per_device_train_batch_size=batch_size,
    gradient_accumulation_steps= 1,
    per_device_eval_batch_size= 24,
    num_train_epochs= 2,
    logging_steps= 100,
    load_best_model_at_end=True,  # Keep the best model according to the evaluation
    metric_for_best_model="f1_score",
    label_names=["labels"],
    dataloader_drop_last=False,
    max_steps= 1500
)
```

```python
dna_bert_classification_model
```

```python
trainer = Trainer(
    dna_bert_classification_model,
    args_entropy,
    train_dataset= tokenized_datasets_train_entropy,
    eval_dataset= tokenized_datasets_validation_entropy,
    tokenizer= tokenizer_dnabert,
    compute_metrics=compute_metrics_f1_score)
```

```python
train_results = trainer.train()
```
